```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
# Cuál es el nombre del primer superhéroe de la tabla heroes?
query1 = """ 
SELECT * 
FROM heroes_information 
limit 1
"""
```

```python
pysqldf(query1)
# A-Bomb
```

```python
# ¿Cuántas casas publicadoras existen?
query2 = """
SELECT Publisher
FROM heroes_information
GROUP BY Publisher;

"""
```

```python
pysqldf(query2)
# 24
```

```python
# ¿Cuántos superhéroes miden más de 2 metros (200 cms)
query3 = """
SELECT *
FROM heroes_information
WHERE Height > 200

"""
```

```python
pysqldf(query3)
# 59
```

```python
# ¿Cuántos superhéroes son humanos y miden más de 2 metros (200 cms)?
query4 = """
SELECT *
FROM heroes_information
WHERE Height > 200
AND Race = 'Human'

"""
```

```python
pysqldf(query4)
# 12
```

```python
# ¿Cuántos superhéroes pesan más de 100 lbs y menos de 200 lbs?
query5 = """
SELECT *
FROM heroes_information
WHERE Weight
BETWEEN 100 AND 200

"""
```

```python
pysqldf(query5)
# 98
```

```python
# ¿Cuántos superhéroes tienen los ojos de color azul o rojo?
query6 = """
SELECT *
FROM heroes_information
WHERE [Eye color]
IN ('blue', 'red')
"""
```

```python
pysqldf(query6)
# 271
```

```python
# ¿Cuántos superhéroes son Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro?
query7 = """
SELECT *
FROM heroes_information
WHERE Race 
IN ('Human', 'Mutant') 
AND [Hair color] = 'Green'
OR Race = 'Vampire' 
AND [Hair color] = 'Black'
"""
```

```python
pysqldf(query7)
# 5 (respuesta diferente a kahoot pero me salen 5)
```

```python
# ¿Cuál es el nombre del primer superhéroe si ordeno la tabla por raza en orden descendente?
query8 = """
SELECT name
FROM heroes_information
ORDER BY Race
DESC;
"""
```

```python
pysqldf(query8)
# Solomon Grundy
```

```python
# ¿Cuántos superhéroes son de género masculino y cuántas de género femenino?
query9 = """
SELECT Gender, COUNT(*) AS cnt
FROM heroes_information
GROUP BY Gender
"""
```

```python
pysqldf(query9)
# hombres:200 mujeres:505
```

```python
# ¿Cuántas casas publicadoras tienen más de 15 superhéroes?
query10 = """
SELECT Publisher, COUNT(*) AS cnt
FROM heroes_information
GROUP BY Publisher
HAVING cnt > 15
"""
```

```python
pysqldf(query10)
# 4
```
